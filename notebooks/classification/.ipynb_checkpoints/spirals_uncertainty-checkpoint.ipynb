```python
def plot_macest_sklearn_comparison_surface(low_range, up_range ,
                                           macest_model, sklearn_model,
                                          plot_training_data = True):
    
    x1_ne = np.random.uniform(-low_range,low_range,10**4)
    x2_ne = np.random.uniform(-low_range,low_range,10**4)
    points_ne = np.array((x1_ne,x2_ne)).T

    x1_fa = np.random.uniform(-up_range,up_range,10**4)
    x2_fa = np.random.uniform(-up_range,up_range,10**4)
    points_fa = np.array((x1_fa,x2_fa)).T
    
    fig, ax = plt.subplots(ncols = 2, nrows =2,
                       figsize = (16,12),)
    
    h0 = ax[0,0].tricontourf(x1_ne,x2_ne,macest_model.predict_confidence_of_point_prediction(points_ne),
                             vmin=0.,
                             vmax =1,
                             cmap = cm.viridis)

    g0 = ax[0,1].tricontourf(x1_ne,x2_ne, np.amax(sklearn_model.predict_proba(points_ne), axis =1), 
                          vmin = 0.,
                          vmax =1,
                          cmap = cm.viridis)
    
    if plot_training_data :
        ax[0,0].scatter(X_pp_train[:,0],X_pp_train[:,1],
                        c=y_pp_train,
                        cmap = cm.rainbow,
                        edgecolors = 'black')

        ax[0,1].scatter(X_pp_train[:,0],X_pp_train[:,1],
                        c=y_pp_train, alpha = 0.9,
                        edgecolors = 'black',
                        cmap = cm.rainbow)


    h1 = ax[1,0].tricontourf(x1_fa,x2_fa,
                             macest_model.predict_confidence_of_point_prediction(points_fa),
                             vmin = 0.,
                             vmax =1,
                             cmap = cm.viridis)

    ax[1,0].scatter(X_pp_train[:,0],X_pp_train[:,1],
                    c=y_pp_train,
                    edgecolors = 'black', 
                    cmap = cm.get_cmap('rainbow', 2))

    g1 = ax[1,1].tricontourf(x1_fa,x2_fa,
                             np.amax(sklearn_model.predict_proba(points_fa), axis =1),
                             vmin = 0.,
                             vmax = 1.,
                             cmap = cm.viridis)

    ax[1,1].scatter(X_pp_train[:,0] ,X_pp_train[:,1],
                    c=y_pp_train,
                    alpha = 0.9,
                    cmap = cm.get_cmap('rainbow', 2),
                    edgecolors = 'black')
    
    fig.colorbar(h1,
                 ax=ax.ravel().tolist(), 
                 ticks= np.arange(0.,1.01, 0.1 ), 
                 label = 'Confidence')
    
    
    ax[0,0].set_title('MACE')
    ax[1,0].set_title('MACE')
    
    ax[0,1].set_title('Random Forest')
    ax[1,1].set_title('Random Forest')
        
    ax[0,0].set_xlabel('x1', fontsize = 18)
    ax[0,1].set_xlabel('x1', fontsize = 18)

    ax[0,0].set_ylabel('x2', fontsize = 18)
    ax[0,1].set_ylabel('x2', fontsize = 18)

    ax[1,0].set_xlabel('x1', fontsize = 18)
    ax[1,1].set_xlabel('x1', fontsize = 18)

    ax[1,0].set_ylabel('x2', fontsize = 18)
    ax[1,1].set_ylabel('x2', fontsize = 18)
```

```python
def create_two_spirals(r, nrotations, n_points, noise):
    theta = np.linspace(0, 2 * np.pi *nrotations , n_points)
    r0 = r * theta
    x_20 = r0*np.cos(theta) + np.random.normal(0, noise, n_points)
    y_20 = r0*np.sin(theta) + np.random.normal(0, noise, n_points)
    
    r1 = r * theta
    x_21 =  - r1*np.cos(theta )+ np.random.normal(-0.4
                                                  , noise, n_points)
    y_21 =  - r1*np.sin(theta ) + np.random.normal(0.4, noise, n_points)
    
    return x_20, x_21, y_20, y_21
```

```python
x_20, x_21, y_20, y_21 = create_two_spirals(0.5, 0.5, 3 * 10**4, 0.15)
```

```python
X1 = np.hstack((x_20, x_21))
X2 = np.hstack((y_20, y_21))
X = np.array((X1, X2)).T
y = np.hstack((np.zeros(3 *10**4), np.ones(3 *10**4))).astype(int)
```

```python
plt.title('training set')
plt.plot(X[y==0,0], X[y==0,1], '.', label='class 1')
plt.plot(X[y==1,0], X[y==1,1], '.', label='class 2')
plt.legend()
plt.show()
```

```python
X_pp_train, X_conf_train, y_pp_train, y_conf_train  = train_test_split(X, y, test_size=0.6, random_state=0)
X_conf_train, X_cal, y_conf_train, y_cal = train_test_split(X_conf_train, y_conf_train,
                                                            test_size=0.5, random_state=0)
X_cal, X_test, y_cal,  y_test, = train_test_split(X_cal, y_cal, test_size=0.5, random_state=0)
```

```python
model = RandomForestClassifier(random_state =0,
                               n_estimators =800)

model.fit(X_pp_train, y_pp_train)
```

```python
macest_model = clmod.ModelWithConfidence(model,
                                      X_conf_train,
                                      y_conf_train)

macest_model.fit(X_cal, y_cal)
```

## What does the surface learnt by the random forest  look like ?

```python
plot_prediction_conf_surface(4, 50, model, True)
```

###Let's Compare with MACE

```python
plot_macest_sklearn_comparison_surface(4, 40, macest_model, model, plot_training_data=True)
```

##Let's now extrapolate the original function

###The predictions effectively have a 50/50 chance of being correct but the Random Forest has a very high confidence for every prediction,This example is clearly quite extreme however it highlights the problem of extrapolating confidence, we do not have any information about this function even a very small distance from the training data

```python
x_20, x_21, y_20, y_21 = create_two_spirals(1, 2, 2 * 10**4, 0.1)
```

```python
X1 = np.hstack((x_20, x_21))
X2 = np.hstack((y_20, y_21))
X = np.array((X1, X2)).T
y = np.hstack((np.zeros(2 *10**4), np.ones(2 *10**4))).astype(int)
```
